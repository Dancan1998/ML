## Matplotlib example workflow

```python
# 1. Prepare some data
x = [1,2,3,4]
y=[11,22,45,66]

# 2. Setup plot / figure
fig, ax = plt.subplots(figsize=(8,8))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save 
fig.savefig("images/sample-plot.png")
```

## Making figures with numpy arrays

```python
# Create some data
x = np.linspace(0,10,100)
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
# Use same data to make scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# Another scatter plot using sin
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
# Make a plot from a dictionary
nut_butter_prices = {"Almond_butter": 10,
                    "Peanut_butter": 8,
                    "Cocoa_butter":12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values());
```

```python
# horizontal bar graph
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

```python
# Make some data for histrogram
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# subplot option 1
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(nrows=2,
                                           ncols=2,
                                           figsize=(10,5))
# Plot to each different axes
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

## New Dataset

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease.head()
```

```python
# Histogram of age
heart_disease["age"].plot.hist(bins=10);
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
heart_disease.head()
```

```python
over_50=heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(kind="scatter", x="age", y="chol", c="target");
```

```python
import matplotlib.pyplot as plt
```

```python
## oo method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter", x="age", y="chol", c="target", ax=ax);
ax.set_xlim([45,100]);
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
ax.set(title="Heart Disease Cholestrol Levels",
      xlabel="Age",
      ylabel="Cholestrol");
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
```
